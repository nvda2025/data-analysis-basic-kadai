# **scikit-learnで予測モデルを変更することで評価結果が変わることを確認しよう**

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# from sklearn.tree import DecisionTreeClassifier
```

```python
# Step1. 8章の「8.4 scikit-learnの基本的な使い方」の「1. サンプルデータの取得」、「2. サンプルデータの分割」で記載の「記述するコード」を記述して実行する。

# 1. サンプルデータの取得
dataset = load_wine()

# 2. サンプルデータの分割
X = dataset.data #成分
y = dataset.target #種類
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
# Step2.「3. 予測モデルのインスタンス化」で記載の「記述するコード」を、「決定木」に対応するコードから「ランダムフォレスト」に対応するコードに変更して実行する。
model = RandomForestClassifier(random_state=3)
```

```python
# Step3.「4. 予測モデルの学習」、「5. 予測モデルの評価」で記載の「記述するコード」を記述して実行する。

# 4. 予測モデルの学習
model.fit(X_train, y_train)

# 5. 予測モデル（ランダムフォレスト）の評価（testデータ54とpredデータの一致率を見る）
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)

# 決定木の一致率（正解率）
0.9259259259259259
```
