# **カテゴリー列の要素の出現頻度をカウントして、その結果を棒グラフで表示する**
# **商品番号列の商品番号ごとに、注文数の基本統計量を求める**

```python
# Step1 ライブラリのインポート
!pip install japanize_matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib
```

```python
df = pd.read_csv("sample_pandas_6.csv", encoding='shift-jis')
category_df = pd.read_csv('category.csv')
df2 = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
df2
```

```python
# Step2 カテゴリー列の要素(弁当・デザートなど)のユニークな出現頻度をカウントするメソッドと、そのメソッドの実行結果を棒グラフ化するメソッドを組み合わせたコードを記述する
# Step3 グラフを表示させる行「plt.show()」を書く
df2['カテゴリー'].value_counts().plot(kind='bar')
plt.title('カテゴリー別出現頻度')  # グラフのタイトルを設定
plt.xlabel('カテゴリー')  # x軸のラベルを設定
plt.ylabel('出現頻度')  # y軸のラベルを設定
plt.show()
```

```python
# Step4 ある基準列(商品番号列)ごとに要素をグループにまとめサブ列を集計するメソッドと、
# 基本統計量を算出するメソッドを組み合わせたコードを記載する
df.groupby('商品番号')['注文数'].sum()
df.groupby('商品番号')['注文数'].describe()
```
